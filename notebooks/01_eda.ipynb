```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
FILE_PATH = "../data/raw/games.csv"
df = pd.read_csv(FILE_PATH)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
d1 = df.isnull().sum().sort_values(ascending=False)
d2 = df.shape[0]
d3 = (d1/d2)*100
d3
```

```python
missing_pct = d3[d3 > 0]
missing_pct
```

```python
colours = ["red" if val > 60 else "orange" if val > 30 else "green" 
           for val in missing_pct.values]

plt.figure(figsize=(10, 6))
sns.barplot(x=missing_pct.values, y=missing_pct.index, hue=colours)
plt.ylabel("Column Names")
plt.xlabel("Percentage values")
plt.title("Zero valued Columns in Percentage")
plt.show()
```

 This chart is used to show which columns have the most Null/missing or Zero values. Using this chart we can drop the columns that provide no analytical value in our analysis.

 The columns in Red can be dropped immediately because over 60% of it contains only 0 values.

 The columns in orange can be kept for further analysis but care needs to be kept when using them.
 
 The columns in green have the fewest zeroes in them so they can be kept for analysis. 

```python
plt.figure(figsize=(10, 6))
plt.barh(y=missing_pct.index, width=missing_pct.values, color=colours)
plt.ylabel("Column Names")
plt.xlabel("Percentage values")
plt.title("Missing Value Analysis by Column")
plt.show()
```
